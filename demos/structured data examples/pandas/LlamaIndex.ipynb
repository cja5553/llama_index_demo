```python
df=pd.read_csv("greater_st_louis_housing_with_density.csv")
df["vacant_unit_rate"]=df["vacant_units"]/df["total_units"]
df
```

## Part II: Index your data
(You can also define your own storage and service index here)

```python
index = GPTPandasIndex(df=df)
```

## Part III: Query the data

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What is the correlation between density and vacant unit rates?")
print(response.response)
```

```python
response = query_engine.query("For each county, what is the number of vacant units per population, sort by highest to lowest")
print(response.response)
```
